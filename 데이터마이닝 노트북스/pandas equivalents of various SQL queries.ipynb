**pandas (software)**

Not to be confused with PANDAS, the Australian archival management system used for the Pandora Archive.

In computer programming, pandas is a software library written for the Python programming language for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series. It is free software released under the three-clause BSD license. The name is derived from the term "panel data", an econometrics term for data sets that include observations over multiple time periods for the same individuals.

Library features

* DataFrame object for data manipulation with integrated indexing.
* Tools for reading and writing data between in-memory data structures and different file formats.
* Data alignment and integrated handling of missing data.
* Reshaping and pivoting of data sets.
* Label-based slicing, fancy indexing, and subsetting of large data sets.
* Data structure column insertion and deletion.
* Group by engine allowing split-apply-combine operations on data sets.
* Data set merging and joining.
* Hierarchical axis indexing to work with high-dimensional data in a lower-dimensional data structure.
* Time series-functionality: Date range generation[4] and frequency conversion, moving window statistics, moving window linear regressions, date shifting and lagging.
* Provides data filtration.

The library is highly optimized for performance, with critical code paths written in Cython or C.[5] 

Dataframes

Pandas is mainly used for machine learning in form of dataframes. Pandas allow importing data of various file formats such as csv, excel etc. Pandas allows various data manipulation operations such as groupby, join, merge, melt, concatenation as well as data cleaning features such as filling, replacing or imputing null values. 

History

Developer Wes McKinney started working on pandas in 2008 while at AQR Capital Management out of the need for a high performance, flexible tool to perform quantitative analysis on financial data. Before leaving AQR he was able to convince management to allow him to open source the library. 

Another AQR employee, Chang She, joined the effort in 2012 as the second major contributor to the library. 

In 2015, pandas signed on as a fiscally sponsored project of NumFOCUS, a 501(c)(3) nonprofit charity in the United States.

# 1. Selecting data (SELECT)

Selecting data from a dataset is commonly used to retrieve and preview data.

For example: to preview sepal_length, sepal_width, petal_length , and petal_width columns from iris data and display the first 2 records.

In SQL, `SELECT` statement is used to select data from a database. It is done using a comma-separated list of columns you would like to select (or a `*` to select all columns):

```
SELECT sepal_l, sepal_w, petal_l, petal_w 
FROM iris 
LIMIT 2
```

In Pandas, columns selection is done by passing a list of column names to your DataFrame:

````python
iris
````

# 2. Selecting data with conditions (WHERE)

In practice, you often would like to retrieve records that fulfill some specified conditions and that is normally done by a filter.

Single condition

For example, to retrieve all the data with label “virginica”
In SQL, the filter is done via a `WHERE` clause:

```
SELECT * 
FROM iris 
WHERE label='virginica'
```

In Pandas, DataFrames can be filtered in many ways and the most widely used is [boolean indexing](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#boolean-indexing):

````python
iris
````

````python
search = iris[iris['label']=='virginica']
search
````

````python
search.shape
````

````python
search = df[df['label']=='virginica']
search
````

````python
search.shape
````

## Multiple conditions AND

In some cases, you would like to retrieve records that fulfill multiple conditions.

For example, to retrieve all the data with label “virginica” and petal_length ≥ 5.5

In SQL, the `WHERE` clause can be combined with `AND` operator to display records if all the conditions separated by `AND` are true:

```
SELECT * 
FROM iris 
WHERE label='virginica' AND petal_l >= 5.5
```

In Pandas, the equivalent AND conditions can be passed to a DataFrame using `&` operator:

````python
iris
````

````python
search = iris[(iris['label']=='virginica') & 
              (iris['petal_l'] >= 5.5)]
search
````

````python
search.shape
````

````python
search = df[(df['label']=='virginica') & 
              (df['petal_l'] >= 6.0)]
search
````

````python
search.shape
````

````python
# setosa: min(petal_l) = 1.9, min(petal_w) = 0.6
search = df[(df['petal_l'] <= 1.9) & 
              (df['petal_w'] <= 0.6)]
search
````

````python
search.shape
````

## Multiple condition OR

For the cases, you would like to extract records that meet any of the conditions.

For example, to retrieve all the data with either petal_length ≥ 5 or septal_length ≥ 6

In SQL, similar to `AND` operator, `OR` operator can be combined with `WHERE` to display records if any of the conditions separated by `OR` is true:

```
SELECT * 
FROM iris 
WHERE petal_l >= 5 OR sepal_l >= 6
```

In Pandas, the equivalent OR conditions can be passed to a DataFrame using `|` operator:

````python
iris
````

````python
search = iris[(iris['petal_l'] >= 5) | 
              (iris['sepal_l'] >= 6)]
search
````

# 3. Working with missing values (NULL or NaN)

In the real-world, many interesting datasets will have some amount of values missing. Different programming language uses a different approach to indicates missing values. And for database and Pandas:

* Missing value is commonly shown as `NULL` in database
* `NaN`, [a special floating-point value](https://towardsdatascience.com/working-with-missing-values-in-pandas-5da45d16e74), is used in Pandas to indicate a missing value.

## Selecting records with missing value

For some cases, you want to see only the records where a field is with a missing value.

For example, to retrieve all data with missing petal_length

In SQL, we can use IS NULL operator to test for NULL values:

```
SELECT * 
FROM iris 
WHERE petal_l IS NULL
```

In Pandas, method `isna()` can be used to check for missing values:

````python
search = iris[iris['petal_l'].isna()]
search
````

````python
search.shape
````

````python
search = df[df['petal_l'].isna()]
search
````

````python
search.shape
````

## Filter out missing values

For some cases, you want to see only the records having a value.

For example, to retrieve all data with petal_length is not missing

In SQL, `IS NOT NULL` operator can be used to filter out missing values:

```
SELECT * 
FROM iris 
WHERE petal_l IS NOT NULL
```

In Pandas, the equivalent query can be done with the method `notna()`:

````python
search = iris[iris['petal_l'].notna()]
search
````

````python
search.shape
````

````python
search = df[df['petal_l'].notna()]
search
````

````python
search.shape
````

# 4. Sorting result (ORDER BY)

When querying data from a dataset, you might also want the result to be returned with a sorted order.

## In Ascending order

For example, to retrieve all data with petal_l ≥ 4 from iris dataset, sorted by label column in ascending order

In SQL, the ORDER BY keyword is used to sort the result-set in ascending (ASC) order:

```
SELECT * 
FROM iris 
WHERE petal_l >= 4 
ORDER BY label ASC
```

In Pandas, the equivalent query can be done with the method `sort_values()` with parameters `by=’label’` and `ascending=True`:

````python
iris
````

````python
search = iris[iris['petal_l'] >= 4].sort_values(by='label', ascending=True)
search
````

````python
search.shape
````

````python
search = df[df['petal_l'] >= 4].sort_values(by='label', ascending=True)
search
````

````python
search.shape
````

## In Descending order

For example, to retrieve all data with petal_l ≥ 4 from iris dataset, sorted by label column in descending order.

In SQL, the ORDER BY keyword is used to sort the result-set in descending (DESC) order:

```
SELECT * 
FROM iris 
WHERE petal_l >= 4 
ORDER BY label DESC
```

In Pandas, the equivalent query can be done with the method 'sort_values()' with parameters `by=’label’` and 'ascending=False':

````python
iris
````

````python
search = iris[iris['petal_l'] >= 4].sort_values(by='label', ascending=False)
search
````

````python
search.shape
````

````python
search = df[df['petal_l'] >= 4].sort_values(by='label', ascending=False)
search
````

# 5. Updating records (UPDATE)

Sometimes, we will need to modify the existing records in a dataset.

For example: update label to virginica for all the records with sepal_length = 7.0 and sepal_width = 3.2

In SQL, the `UPDATE` statement together with `SET` operator is used to modify the existing records:

```
UPDATE iris 
SET label = 'virginica' 
WHERE sepal_l = 7 AND sepal_w = 3.2
```

In Pandas, `loc()` can be used to access rows conditionally and return the specified column `label`. After that, we can set value for all records to virginica:

````python
iris
````

````python
iris.loc[(iris['sepal_l'] == 7) & 
        (iris['sepal_w'] == 3.2), 'label'] = 'virginica'
iris
````

````python
iris.loc[(iris['sepal_l'] == 7) & 
        (iris['sepal_w'] == 3.2), 'label'] = 'versicolor'
iris
````

# 6. Grouping records (GROUP BY)

For some cases, we would like to group records that have the same values.

For example: find the number of record in each label

In SQL, the GROUP BY statement groups rows that have the same values into summary rows,

```
SELECT label, count(*) 
FROM iris 
GROUP BY label
```

In Pandas, SQL’s `GROUP BY` operations are performed using the similarly named `groupby()` method:

````python
iris
````

````python
search = iris.groupby('label').size()
search
````

## With multiple stats

For the cases where multiple stats are required

For example: to find the average sepal_length and maximum petal_length in each label

In SQL, there are some built-in [math functions](https://www.w3schools.com/sql/sql_ref_sqlserver.asp) we can use to accomplish the task:

```
SELECT label, avg(sepal_l), max(petal_l) 
FROM iris 
GROUP BY label
```

In Pandas, `agg()` can be used to apply multiple functions at once:

````python
iris
````

````python
search = iris.groupby('label').agg(
       { 'sepal_l': np.mean, 'petal_l': np.max})
search
````

# 8. Querying data from multiple datasets (JOIN)

So far, we have been looking at the query from a single dataset. In the real-world, it is common to combine records from two or more datasets based on a related column. Officially, this kind of query is called Join.

There are 4 different types of Joins

1. Inner Join: returns records that have matching values in both datasets
2. Left (Outer) Join: returns all records from the left dataset, and the matched records from the right dataset
3. Right (Outer) Join: returns all records from the right dataset, and the matched records from the left dataset
4. Full (Outer) Join: returns all records when there is a match in either left or right dataset

````python
#  https://www.w3schools.com/sql/sql_join.asp

Image("f:/WORK/Python/JOIN.png") 
````

Before we diving into the details, let’s make up a new dataset country, which has records about where the flower was found:

````python
country = pd.DataFrame({ 'label': ['virginica', 'versicolor'],
                  'country': ['US', 'Germany']})
country
````

Notice that we don’t have a country record for label “setosa”.

## Inner Join

For example: to retrieve all records that have matching values in both iris and country datasets, based on the label:

```
SELECT * FROM iris 
INNER JOIN country 
ON iris.label = country.label
```

In Pandas, the method `merge()` offers parameters for cases when you’d like to join two DataFrames based on a related column. And by default, it does the inner join:

````python
Image("f:/WORK/Python/JOIN.png") 
````

````python
iris
````

````python
country
````

````python
# No label "setosa"
inner = pd.merge(iris, country, on='label')
inner
````

## Left Outer Join

For example: to retrieve all records from the iris, and the matched records from the country, based on the label:

```
SELECT * FROM iris 
LEFT OUTER JOIN country
ON iris.label = country.label
```

The Pandas equivalent:

# 9. Combining the results of multiple queries (UNION)

In data processing, it is also common to combine the results of two or more queries into a single result. In order to combine the results, each query must have the same number of columns with same data types.

For demonstration, let’s make up two datasets from iris. Note, we also make up a duplicate record as shown in the figure:

````python
iris_a = iris[0:3]
iris_a
````

````python
iris_b = iris[2:]
iris_b
````

## To allow duplicate values

In SQL, the UNION ALL operator can be used to allow duplicate values:

```
SELECT * FROM iris_a WHERE label = 'versicolor' 
UNION ALL 
SELECT * FROM iris_b WHERE label = 'versicolor'
```

In Pandas, the equivalent method is `concat()` and by default it allows all the duplicates:

````python
iris_a
````

````python
iris_b
````

````python
union = pd.concat([iris_a[iris_a['label']=='versicolor'], iris_b[iris_b['label']=='versicolor']])
union
````

## Without duplicate values

In SQL, the UNION operator selects only distinct values:

```
SELECT * FROM iris_a WHERE label = 'versicolor' 
UNION 
SELECT * FROM iris_b WHERE label = 'versicolor'
```

In Pandas, you can use 'concat()' in conjunction with 'drop_duplicates()':

````python
iris_a
````

````python
iris_b
````

````python
union = pd.concat([iris_a[iris_a['label']=='versicolor'], iris_b[iris_b['label']=='versicolor']]).drop_duplicates()
union
````
