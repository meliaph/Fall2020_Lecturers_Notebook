&#x1F449; Walter Elias Disney, was born in Chicago, Illinois, on December 5, 1901. 

&#x1F449; He is famous as the creator of **Mickey Mouse** and the founder of **Disneyland** and **Walt Disney Co.**

![](https://assets.entrepreneur.com/content/3x2/2000/20150825195012-walt-disney-land-resort.jpeg?width=700&crop=2:1)

&#x1F449; He is an artist and entrepreneur

&#x1F449; He attended evening classes at the Chicago Academy of Fine Arts in 1917 and hoped to become a newspaper cartoonist.

&#x1F449; He got his first art job at the Pesmen-Rubin Commercial Art Studio in 1919

&#x1F449; He also formed a business with Ub Iwerks named Iwerks-Disney Commercial Artists in 1920 and got him his first experience in animation because he get the chance to collaborate with Kansas City Slide Company.

&#x1F449; He later created Newman Laugh-O-grams, another business that produces advertising and topical shorts and story cartoons.

&#x1F449; In 1922, he incorporated his latest business as Laugh-O-grams Films, Inc. that valued $15,000.

&#x1F449; But in 1923, Laugh-O-grams Films, Inc. went bankrupt and has to made him move to Hollywood. 

&#x1F449; His career in Hollywood started as a director and he later established Disney Brothers Studio with his brother. 

&#x1F449; In 1926, Disney Brothers Studio changed its name and became **THE WALT DISNEY STUDIO** with his brother as co-founder, Iwerks as one of his animator, and Lilian Bounds (his wife) as an inker.

&#x1F449; The birth of **THE WALT DISNEY STUDIO** started here and this studio is located in Hyperion Avenue, which later be known as the birthplace of many of Disney's greatest films.
